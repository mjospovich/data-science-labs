```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, sys, re
import seaborn as sns

episodes = pd.read_csv('simpsons_episodes.csv')
```

```python
# extract some datetime variables
episodes['datetime'] = pd.to_datetime(episodes['original_air_date'])
episodes['datetime_year'] = episodes['datetime'].dt.year
episodes['datetime_month'] =episodes['datetime'].dt.month
episodes['datetime_day'] =episodes['datetime'].dt.day
episodes['datetime-MM-DD'] = episodes['datetime'].dt.strftime('%m-%d')
```

## Vizualizacija dviju kontinuiranih varijabli

- Je li se smanjivao IMDB rating i US views kroz godine
- Postoji mnogo načina kako se može ispitati odnos između dvije varijable. Najjednostavniji je dijagram scatterplot, koristeći plt.scatterplot()

```python
plt.figure(figsize=(20, 7))

sns.scatterplot(x='datetime_year', y='imdb_rating', size='us_viewers_in_millions', data=episodes, alpha=0.7)
plt.title('IMDB rating and US viewers through the years')
```

## Vizualizacija dviju varijabli, kontinuiranih i kategoričkih

- Razmotrimo IMDB rating i US views kroz godine i sezone emitiranja
- Prikazat ćemo barplot, boxplot i violinplot. Seaborn nudi mnogo više uvjerljivih vizualizacija. Bacite oko na primjere: https://seaborn.pydata.org/examples/index.html

```python
# barplot
# make a figure that contains two seperate barplots that on x has years on both and ond one y is views on other y is rating“
plt.figure(figsize=(20, 12))
plt.subplot(2, 1, 1)
sns.barplot(x='season', y='us_viewers_in_millions', data=episodes, palette='viridis')
plt.title('US viewers through the seasons')
plt.subplot(2, 1, 2)
sns.barplot(x='datetime_year', y='imdb_rating', data=episodes, palette='viridis')
plt.title('IMDB rating through the years')

```

```python
# boxplot
# plt.figure(figsize=(20, 12))
# plt.subplot(2, 1, 1)
# sns.barplot(x='season', y='us_viewers_in_millions', data=episodes, palette='viridis')
# plt.title('US viewers through the seasons')
# plt.subplot(2, 1, 2)
# sns.barplot(x='datetime_year', y='imdb_rating', data=episodes, palette='viridis')
# plt.title('IMDB rating through the years')
# redo this using boxplot

plt.figure(figsize=(20, 12))
plt.subplot(2, 1, 1)
sns.boxplot(x='season', y='us_viewers_in_millions', data=episodes, palette='spring')
plt.title('US viewers through the seasons')
plt.subplot(2, 1, 2)
sns.boxplot(x='datetime_year', y='imdb_rating', data=episodes, palette='autumn')
plt.title('IMDB rating through the years')

```

```python
# violinplot
plt.figure(figsize=(20, 12))
plt.subplot(2, 1, 1)
sns.violinplot(x='season', y='us_viewers_in_millions', data=episodes, palette='husl')
plt.title('US viewers through the seasons')
plt.subplot(2, 1, 2)
sns.violinplot(x='datetime_year', y='imdb_rating', data=episodes, palette='rainbow')
plt.title('IMDB rating through the years')
```

## 2.1 Heatmape

- Vizualizirajte IMDB i US views kroz sezone te koristite funkciju Seaborn heatmap te funkciju *pivot* of pandasa

- Konačno, heatmapom možemo predstaviti tri varijable, kodiranjem treće varijable bojom. Pokažite kako sa sezonom opada IMDB ocjena i US views korištenjem heatmape


```python
# Napišite svoj kod ovdje da biste kreirali prvu heatmapu
# create a heatmap that shows all episodes throgh the seasons and their ratings, use pandas pivot_table
# keys for csv: id,title,original_air_date,production_code,season,number_in_season,number_in_series,us_viewers_in_millions,views,imdb_rating,imdb_votes,image_url,video_url

episodes_pivot = episodes.pivot_table(index='number_in_season', columns='season', values='imdb_rating')
plt.figure(figsize=(20, 12))
sns.heatmap(episodes_pivot, cmap='coolwarm', linecolor='white',
             annot=True, linewidths=3,
             cbar_kws={'label': 'IMDB rating'})
plt.title('IMDB rating of episodes through the seasons')
plt.xlabel('Season')
plt.ylabel('Episode number in season')


```

```python
# Napišite svoj kod ovdje da biste kreirali drugu heatmapu
episodes_pivot = episodes.pivot_table(index='number_in_season', columns='season', values='us_viewers_in_millions')
plt.figure(figsize=(20, 12))
sns.heatmap(episodes_pivot, cmap='coolwarm', linecolor='white',
             annot=True, linewidths=3,
             cbar_kws={'label': 'US viewers in millions'})
plt.title('US views in millions of episodes through the seasons')
plt.xlabel('Season')
plt.ylabel('Episode number in season')
```
